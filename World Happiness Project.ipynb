## 2) EDA (Exploratory Data Analysis):

```python
happiness.shape
```

```python
#separating the numerical and categorical columns
num_columns=[]
cat_columns=[]
for i in happiness:
    if happiness.dtypes[i]=="object":
        cat_columns.append(i)
    else:
        num_columns.append(i)
```

```python
#checking the categorical columns
cat_columns
```

```python
#checking the numerical columns
num_columns
```

```python
len(num_columns)
```

```python
#lets see how data is distributed for every column
plt.figure(figsize=(35,20),facecolor='pink')
plotnumber=1
for column in num_columns:
    if plotnumber<11:
        ax=plt.subplot(2,5,plotnumber)
        sns.distplot(happiness[column])
        plt.xlabel(column,fontsize=30)
    plotnumber+=1
plt.tight_layout()
```

we observe that there are outliers at
standard error,
family,
generosity,
dtstopia residual columns in tha dataset but for the clarify purpose we have to draw box plot for checking outliers

```python
#visualizing the outliers using Box Plot
plt.figure(figsize=(20,15),facecolor="orange")
plotno=1
for col in num_columns:
    if plotno<=10:
        plt.subplot(2,5,plotno)
        sns.boxplot(happiness[col])
        plt.xlabel(col,fontsize=15)
    plotno+=1
plt.show()
```

yes there are outliers are present in dataset at standard error,
family,
trust,
generosity,
dtstopia residual columns in tha dataset. we have to delete those outliers.

```python
#1st Quantile
q1=happiness.quantile(0.25)
print("FIRST QUANTILE\n")
print(q1)
print("\n")
 

#3rd Quantile
q3=happiness.quantile(0.75)
print("THIRD QUANTILE\n")
print(q3)
print("\n")

#find the IQR (Inter Quantile Range) to identify outliers
# IQR = Q3 - Q1
iqr=q3-q1
print("INTER QUANTILE RANGE\n")
print(iqr)
```

```python
#validating upperside outlier
stderror_high=(q3.Standard Error(1.5*iqr.Standard Error))
print(stderror_high)

#check the indexes which have higher values
ind=np.where(happiness["Standard_Error"]>stderror_high)
print(ind)

#drop the indexes which having high values
happiness=happiness.drop(happiness.index[ind])

#shape of the dataset
happiness.shape

#reset indexes
happiness.reset_index()
```
